```python
import requests

def get_related_concepts(start_term):
    # ConceptNet API 엔드포인트 URL
    url = "http://api.conceptnet.io/c/ko/"+ start_term + "?"

    # API 호출
    response = requests.get(url).json()

    # 관련 개념 정보 추출
    related_concepts = []
    for edge in response["edges"]:
        end_term = edge["end"]["label"]
        related_concepts.append(end_term)

    return related_concepts
```

```python
start_term = "자동차"

related_concepts = get_related_concepts(start_term)
related_concepts = set(related_concepts)

print("Related Concepts:")
for concept in related_concepts:
    print(concept)

print()
print(len(related_concepts))
```
